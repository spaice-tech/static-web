```python
# All the images in ./src/assets/images/grid.nosync/ whose filename is not s<id>.jpg or r<id>.jpg will be renamed to s<id>.jpg, where <id> is the first number not used in the current directory.

import os
import re
import shutil

def rename_images():
    path = './src/assets/images/grid.nosync/'
    files = os.listdir(path)
    files.sort()
    i = 11
    for file in files:
        if re.match(r's\d+.jpg', file) or re.match(r'r\d+.jpg', file):
            continue
        else:
            print(file)
            shutil.move(path + file, path + 's' + str(i) + '.jpg')
            i += 1
```

```python
rename_images()
```

```python
import random
# rename the images in ./src/assets/images/grid.nosync/ to random ids if original id is higher than 10
def randomise_images():
    path = './src/assets/images/grid.nosync/'
    files = os.listdir(path)
    original_ids = []
    for file in files:
        #find id with regex
        if len(re.findall(r'\d+', file)) == 0:
            continue
        id = int(re.findall(r'\d+', file)[0])
        if id > 10:
            original_ids.append(id)
    
    # check which ids are missing
    missing_ids = []
    for i in range(11, 100):
        if i not in original_ids:
            missing_ids.append(i)
    print(missing_ids)

    #randomise the ids
    random.shuffle(original_ids)

    # save to temp folder
    i = 0
    for file in files:

        if len(re.findall(r'\d+', file)) == 0:
            continue
        
        id = int(re.findall(r'\d+', file)[0])
        if id > 10:
            shutil.move(path + file, path + 'temp/' + 's' + str(original_ids[i]) + '.jpg')
            i += 1

    # move back to original folder
    files = os.listdir(path + 'temp/')
    for file in files:
        shutil.move(path + 'temp/' + file, path + file)
        
```

```python
randomise_images()
```
